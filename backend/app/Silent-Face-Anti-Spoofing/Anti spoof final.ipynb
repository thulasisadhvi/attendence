```python
from cvzone.FaceDetectionModule import FaceDetector
import cv2
import cvzone  # Make sure cvzone is imported for putTextRect and cornerRect

# Initialize video capture
cap = cv2.VideoCapture(0)

# Check if the webcam is working
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

# Initialize the face detector
detector = FaceDetector(minDetectionCon=0.5, modelSelection=0)

while True:
    success, img = cap.read()

    # Ensure a frame is successfully captured
    if not success or img is None:
        print("Failed to grab frame.")
        break

    # Detect faces
    img, bboxs = detector.findFaces(img, draw=False)

    if bboxs:
        for bbox in bboxs:
            center = bbox["center"]
            x, y, w, h = bbox['bbox']
            score = int(bbox['score'][0] * 100)

            # Draw details
            cv2.circle(img, center, 5, (255, 0, 255), cv2.FILLED)
            cvzone.putTextRect(img, f'{score}%', (x, y - 10))
            cvzone.cornerRect(img, (x, y, w, h))

    # Show the frame (outside the bbox loop)
    cv2.imshow("Image", img)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```

```python
import math
import time
import cv2
import cvzone
from ultralytics import YOLO

# Set confidence threshold
confidence_threshold = 0.6

# Initialize webcam
cap = cv2.VideoCapture(0)
cap.set(3, 640)  # Width
cap.set(4, 480)  # Height

# Load custom YOLO model
model = YOLO("l_version_1_300.pt")

# Define your class names in order
classNames = ["fake", "real"]

# Frame timing for FPS display
prev_frame_time = 0

while True:
    # Capture frame
    success, img = cap.read()
    if not success:
        print("❌ Failed to grab frame from webcam.")
        break

    # Current frame time for FPS calculation
    new_frame_time = time.time()

    # Run YOLO model
    results = model(img, stream=True, verbose=False)

    for r in results:
        boxes = r.boxes
        for box in boxes:
            # Get coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            w, h = x2 - x1, y2 - y1

            # Get confidence
            conf = float(box.conf[0])

            # Get class index
            cls = int(box.cls[0])

            # Only show results above confidence threshold
            if conf > confidence_threshold:
                label = classNames[cls]
                color = (0, 255, 0) if label == 'real' else (0, 0, 255)

                # Draw rectangle and label
                cvzone.cornerRect(img, (x1, y1, w, h), colorC=color, colorR=color)
                cvzone.putTextRect(
                    img,
                    f'{label.upper()} {int(conf * 100)}%',
                    (max(0, x1), max(35, y1)),
                    scale=2,
                    thickness=3,
                    colorR=color,
                    colorB=color
                )

    # Calculate and display FPS
    fps = 1 / (new_frame_time - prev_frame_time)
    prev_frame_time = new_frame_time
    cv2.putText(img, f'FPS: {int(fps)}', (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 255, 255), 2)

    # Show image
    cv2.imshow("YOLO Real vs Fake Detection", img)

    # Exit on 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import math

# Load YOLO model (trained on COCO dataset)
model = YOLO("yolov8n.pt")  # Or use "yolov8s.pt" if you want better accuracy

# Class labels
PHONE_LABELS = ["cell phone"]
PERSON_LABEL = "person"

# Initialize webcam
cap = cv2.VideoCapture(0)

def get_coords(box):
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    return (x1, y1, x2, y2)

def is_inside(inner, outer):
    ix1, iy1, ix2, iy2 = inner
    ox1, oy1, ox2, oy2 = outer
    return ix1 >= ox1 and iy1 >= oy1 and ix2 <= ox2 and iy2 <= oy2

while True:
    ret, frame = cap.read()
    if not ret:
        break

    detections = model(frame, stream=True)
    phones, faces = [], []

    for r in detections:
        for box in r.boxes:
            cls = int(box.cls[0])
            label = model.names[cls]
            coords = get_coords(box)

            if label in PHONE_LABELS:
                phones.append(coords)
                cv2.rectangle(frame, coords[:2], coords[2:], (0, 0, 255), 2)
                cv2.putText(frame, "Phone", (coords[0], coords[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)

            elif label == PERSON_LABEL:
                faces.append(coords)
                cv2.rectangle(frame, coords[:2], coords[2:], (0, 255, 0), 2)
                cv2.putText(frame, "Face", (coords[0], coords[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    for face in faces:
        for phone in phones:
            if is_inside(face, phone):
                cv2.putText(frame, "SPOOF DETECTED!", (face[0], face[1] - 40), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 255), 3)

    cv2.imshow("Anti-Spoof YOLO", frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO("yolov8n.pt")  # Use a better model if needed

PHONE_LABELS = ["cell phone"]
PERSON_LABEL = "person"

cap = cv2.VideoCapture(0)

def get_coords(box):
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    return (x1, y1, x2, y2)

def is_inside(inner, outer):
    ix1, iy1, ix2, iy2 = inner
    ox1, oy1, ox2, oy2 = outer
    return ix1 >= ox1 and iy1 >= oy1 and ix2 <= ox2 and iy2 <= oy2

while True:
    ret, frame = cap.read()
    if not ret:
        break

    detections = model(frame, stream=True)
    phones, faces = [], []
    spoof_detected = False

    for r in detections:
        for box in r.boxes:
            cls = int(box.cls[0])
            label = model.names[cls]
            coords = get_coords(box)

            if label in PHONE_LABELS:
                phones.append(coords)
            elif label == PERSON_LABEL:
                faces.append(coords)

    for face in faces:
        for phone in phones:
            if is_inside(face, phone):
                spoof_detected = True
                break
        if spoof_detected:
            break

    # Display spoof status in top-left corner
    if spoof_detected:
        cv2.putText(frame, "SPOOF ATTEMPT!", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 255), 3)
    else:
        cv2.putText(frame, "REAL", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 3)

    cv2.imshow("Spoof Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import os

def list_all_files(base_path, indent=0):
    try:
        items = os.listdir(base_path)
    except Exception as e:
        print("Error reading directory:", e)
        return

    for item in items:
        full_path = os.path.join(base_path, item)
        print("    " * indent + f"- {item}/" if os.path.isdir(full_path) else "    " * indent + f"- {item}")
        if os.path.isdir(full_path):
            list_all_files(full_path, indent + 1)

# Set path to your cloned repo folder
repo_path = "Silent-Face-Anti-Spoofing"

print(f"\n📁 Contents of: {repo_path}\n")
list_all_files(repo_path)
```
